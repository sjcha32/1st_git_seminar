```python
import requests
import json
r = requests.get('https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_municipalities_geo_simple.json')
c = r.content
geo = json.loads(c)
```

```python
eat_busan_1 = eat_busan.loc[eat_busan['분류'] == '한식']
eat_busan_2 = eat_busan.loc[eat_busan['분류'] == '카페/찻집']
eat_busan_3 = eat_busan.loc[eat_busan['분류'] == '간이음식']
eat_busan_4 = eat_busan.loc[eat_busan['분류'] == '음식점기타']
```

```python
see_busan_1 = see_busan.loc[see_busan['분류'] == '교통시설']
see_busan_2 = see_busan.loc[see_busan['분류'] == '백화점']
see_busan_3 = see_busan.loc[see_busan['분류'] == '쇼핑몰']
see_busan_4 = see_busan.loc[see_busan['분류'] == '전시시설']
see_busan_5 = see_busan.loc[see_busan['분류'] == '시장']
see_busan_6 = see_busan.loc[see_busan['분류'] == '대형마트']
see_busan_7 = see_busan.loc[see_busan['분류'] == '도시공원']
see_busan_8 = see_busan.loc[see_busan['분류'] == '테마공원']
```

```python
from folium.plugins import MarkerCluster

m = folium.Map(
    location=[35.220264, 129.086444],
    zoom_start=12, 
    tiles='cartodbpositron'
)

folium.GeoJson(
    geo,
    name='지역구'
).add_to(m)

marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(eat_busan_1['위도'], eat_busan_1['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)

for lat, long in zip(eat_busan_2['위도'], eat_busan_2['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="blue")).add_to(marker_cluster)

for lat, long in zip(eat_busan_3['위도'], eat_busan_3['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="black")).add_to(marker_cluster)

for lat, long in zip(eat_busan_4['위도'], eat_busan_4['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="red")).add_to(marker_cluster)

for lat, long in zip(see_busan_1['위도'], see_busan_1['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="gray")).add_to(marker_cluster)

for lat, long in zip(see_busan_2['위도'], see_busan_2['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="lightred")).add_to(marker_cluster)

for lat, long in zip(see_busan_3['위도'], see_busan_3['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="lightred")).add_to(marker_cluster)

for lat, long in zip(see_busan_4['위도'], see_busan_4['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="lightblue")).add_to(marker_cluster)

for lat, long in zip(see_busan_5['위도'], see_busan_5['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="pink")).add_to(marker_cluster)

for lat, long in zip(see_busan_6['위도'], see_busan_6['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="pink")).add_to(marker_cluster)

for lat, long in zip(see_busan_7['위도'], see_busan_7['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="lightgreen")).add_to(marker_cluster)

for lat, long in zip(see_busan_8['위도'], see_busan_8['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="lightgreen")).add_to(marker_cluster)

m
```
